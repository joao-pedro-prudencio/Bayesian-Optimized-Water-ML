# LightGBM

```python
model_lgbm = LGBMRegressor(
    bagging_fraction = 0.2,
    feature_fraction = 0.9,
    learning_rate = 0.15,
    max_depth = 6,
    n_estimators = 800,
    num_leaves = 7,
    reg_alpha = 0.7,
    reg_lambda = 1
)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

def calcular_metricas_regressao(y_test, y_pred):

    rmse = mean_squared_error(y_test, y_pred) ** 0.5
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

    metricas = {
        'Raiz do Erro Quadrático Médio': round(rmse, 4),
        'Erro Absoluto Médio': round(mae, 4),
        'R2 Score': round(r2, 4),
        'Erro Percentual Absoluto Médio (MAPE)': round(mape, 4)
    }

    return metricas
```

```python
%%time

model_lgbm.fit(X_train, y_train)
y_pred_lgbm = model_lgbm.predict(X_test)

calcular_metricas_regressao(y_test, y_pred_lgbm)
```

```python
from yellowbrick.regressor import prediction_error

visualizer = prediction_error(model_lgbm, X_train, y_train, X_test, y_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score

# Predição
y_pred = model_lgbm.predict(X_test)
r2 = r2_score(y_test, y_pred)

# Gráfico
plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred, alpha=0.5, edgecolor='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', label='Ideal')
plt.xlabel("Valor real")
plt.ylabel("Valor predito")
plt.title("Erro de predição - LightGBM")
plt.legend()
plt.grid(True)

# R² no canto superior esquerdo
plt.text(
    0.03, 0.92,
    f"R² = {r2:.3f}",
    transform=plt.gca().transAxes,
    fontsize=12,
    verticalalignment='top',
    bbox=dict(boxstyle="round", facecolor="white", edgecolor="gray", alpha=0.8)
)

plt.tight_layout()
plt.show()
```

```python
from yellowbrick.model_selection import FeatureImportances

viz = FeatureImportances(model_lgbm, relative=False)
viz.fit(X_train, y_train)
viz.show()
```

```python
import matplotlib.pyplot as plt

# Importâncias
importances = model_lgbm.feature_importances_
features = X_train.columns

# Plot manual
pd.Series(importances, index=features).sort_values().plot(kind='barh', figsize=(10,6))
plt.title("Importância das variáveis (LightGBM)")
plt.grid(True)
plt.show()
```

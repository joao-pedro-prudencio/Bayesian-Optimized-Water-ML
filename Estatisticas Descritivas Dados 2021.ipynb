```python
# Importando bibliotecas necessárias
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

X = add_constant(df_filtrado_numerico)

vif_data = pd.DataFrame()
vif_data['Variáveis'] = X.columns

vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

```python
y = df_filtrado_numerico['IN049_AE - Índice de perdas na distribuição']
y
```

```python
X = df_filtrado_numerico.drop('IN049_AE - Índice de perdas na distribuição', axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

import statsmodels.api as sm

X_train_com_constante = sm.add_constant(X_train)

X_train_com_constante

modelo_statsmodels = sm.OLS(y_train, X_train_com_constante, hasconst = True).fit()
```

```python
# F-statistic : significancia conjunta
# t : significancia individual 
# Ver se Prob (F-statistic) é acima de 0.05, não pode ser
```

```python
print(modelo_statsmodels.summary())
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def calcular_metricas_regressao(y_test, y_pred):

    rmse = mean_squared_error(y_test, y_pred) ** 0.5
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

    metricas = {
        'Raiz do Erro Quadrático Médio': round(rmse, 4),
        'Erro Absoluto Médio': round(mae, 4),
        'R2 Score': round(r2, 4),
        'Erro Percentual Absoluto Médio (MAPE)': round(mape, 4)
    }

    return metricas
```

```python
y_previsto = modelo.predict(X_test)

calcular_metricas_regressao(y_test, y_previsto)
```

# Floresta Aleatória

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

def calcular_metricas_regressao(y_test, y_pred):

    rmse = mean_squared_error(y_test, y_pred) ** 0.5
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

    metricas = {
        'Raiz do Erro Quadrático Médio': round(rmse, 4),
        'Erro Absoluto Médio': round(mae, 4),
        'R2 Score': round(r2, 4),
        'Erro Percentual Absoluto Médio (MAPE)': round(mape, 4)
    }

    return metricas
```

```python
%%time

model_rf.fit(X_train, y_train)
y_pred_rf = model_rf.predict(X_test)

calcular_metricas_regressao(y_test, y_pred_rf)
```

```python
from yellowbrick.regressor import prediction_error

visualizer = prediction_error(model_rf, X_train, y_train, X_test, y_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score

# Predição
y_pred = model_rf.predict(X_test)
r2 = r2_score(y_test, y_pred)

# Gráfico
plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred, alpha=0.5, edgecolor='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', label='Ideal')
plt.xlabel("Valor real")
plt.ylabel("Valor predito")
plt.title("Erro de predição - Random Forest")
plt.legend()
plt.grid(True)

# R² no canto superior esquerdo
plt.text(
    0.03, 0.92,
    f"R² = {r2:.3f}",
    transform=plt.gca().transAxes,
    fontsize=12,
    verticalalignment='top',
    bbox=dict(boxstyle="round", facecolor="white", edgecolor="gray", alpha=0.8)
)

plt.tight_layout()
plt.show()
```
